```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Units": [120, 150, 170, 160, 180, 200]
}

df = pd.DataFrame(data)
df

rate = 5
df["Bill"] = df["Units"] * rate
df

plt.plot(df["Month"], df["Units"], marker="o")
plt.title("Monthly Electricity Usage")
plt.xlabel("Month")
plt.ylabel("Units Consumed")
plt.show()

plt.bar(df["Month"], df["Bill"])
plt.title("Monthly Electricity Bill")
plt.xlabel("Month")
plt.ylabel("Bill (₹)")
plt.show()

print("Total Units:", df["Units"].sum())
print("Average Units:", df["Units"].mean())
print("Maximum Units:", df["Units"].max())
print("Minimum Units:", df["Units"].min())

next_month = np.mean(df["Units"])
print("Predicted Units for Next Month:", next_month)

df.to_csv("electricity_bill.csv", index=False)

max_month = df.loc[df["Units"].idxmax()]
print("Highest usage month:", max_month["Month"])
```
