```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Charger"],
    "Sales": [50, 120, 40, 90, 70]
}

df = pd.DataFrame(data)
df

print("Total Sales:", np.sum(df["Sales"]))
print("Average Sales:", np.mean(df["Sales"]))

max_product = df.loc[df["Sales"].idxmax()]
print("Most Demanded Product:", max_product["Product"])

plt.bar(df["Product"], df["Sales"])
plt.title("Product Demand Analysis")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()

plt.pie(df["Sales"], labels=df["Product"], autopct="%1.1f%%")
plt.title("Product Demand Share")
plt.show()

data2 = {
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Laptop": [10, 15, 12, 13],
    "Mobile": [30, 35, 28, 27]
}

df2 = pd.DataFrame(data2)

plt.plot(df2["Month"], df2["Laptop"], marker="o", label="Laptop")
plt.plot(df2["Month"], df2["Mobile"], marker="o", label="Mobile")

plt.title("Product Demand Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

```
